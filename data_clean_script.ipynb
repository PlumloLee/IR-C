The Single-Family Loan-Level Dataset is split in calendar quarters, beginning with the first quarter of 1999 and 
ending with the quarter of the Origination Cutoff Date. For each calendar quarter, there is


- One Origination Data file containing loan-level origination information for all the loans originated during 
the quarter. 


- One Performance Data file for all of the respective loans originated during the quarter.
    - All performance periods associated to a loan will be contained within the same Performance 
Data file.
    - Example: a loan originated in January 1999 which has not been assigned a Zero Balance Code 
as of January 2019 will have its 201901 period found in the 1999Q1 Performance Data file

## Merging the dataset

Here we only focus on the 30 year fixed coupon loans only.

```python
print(data['Original Loan Term'].value_counts())
print(data['Amortization Type (Formerly Product Type)'] .value_counts())
# filter out 30 year fixed rate mortgages
data_30yfix = data[(data['Amortization Type (Formerly Product Type)'] == 'FRM') & (data['Original Loan Term'] == 360)]
```

```python
# merge the two datasets
data_merged = pd.merge(data_30yfix, data2, on='Loan Sequence Number', how='inner')

# filter out the columns that are not needed
# data_merged = data_merged[['Loan Sequence Number', 'Monthly Reporting Period', 'Current Actual UPB', 'Current Loan Delinquency Status',
#                            'Loan Age', 'Remaining Months to Legal Maturity', 'Repurchase Flag', 'Modification Flag', 'Zero Balance Code',
#                            'Zero Balance Effective Date', 'Current Interest Rate']]


# convert the 'Current Interest Rate' to float
data_merged['Current Interest Rate'] = data_merged['Current Interest Rate'].astype(float)

# convert the 'Monthly Reporting Period' to datetime
data_merged['Monthly Reporting Period'] = pd.to_datetime(data_merged['Monthly Reporting Period'], format='%Y%m')

# sort the data by 'Loan Sequence Number' and 'Monthly Reporting Period'
data_merged = data_merged.sort_values(by=['Loan Sequence Number', 'Monthly Reporting Period'])

# reset the index
data_merged = data_merged.reset_index(drop=True)

data_merged.head()
```

# Detecting Prepayment

Mortgages are exposed to two types of event risk: prepayment and default. Part of the challenge is to identify which of the two events has affected a particular loan. Here are some suggestions:


(i) Initially, focus on the prepayment risk only.

(ii) Do not treat curtailments as prepayments, only repayments of the total outstanding balance should be treated as prepayments.

```python
# Detecting Prepayment

# if the zero balance code is 1 , then the loan is paid off or matured. 
# if REMAINING MONTHS TO LEGAL MATURITY is 0, then the loan is matured.
# prepayment is where ['Zero Balance Code'] == 1 & (data_merged['Remaining Months to Legal Maturity'] > 0)
data_merged['Prepayment'] = (data_merged['Zero Balance Code'] == 1) & (data_merged['Remaining Months to Legal Maturity'] > 0)

# count the number and ratio of prepayments
prepayment_count = data_merged['Prepayment'].sum()
prepayment_ratio = prepayment_count / len(data_merged)
prepayment_count, prepayment_ratio
```
